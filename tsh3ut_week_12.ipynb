## Pie Chart (3 points)

```python
# Create a pie chart from raw values
import matplotlib.pyplot as plt

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
colors = ['c', 'g', 'y']
```

```python
plt.pie(slices, labels=labels, colors=colors,
        startangle=90,
        autopct = '%1.2f%%')

plt.axis('equal')
plt.title('Pie Chart from Raw Values')
plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)
```

```python
m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Raw Data")
plt.show()

plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
plt.title('Scatter Plot with Trendline')
z = np.polyfit(x, y, 2)
p = np.poly1d(z)
plt.plot(x, p(x))
plt.scatter(x,y)
plt.show()
```
